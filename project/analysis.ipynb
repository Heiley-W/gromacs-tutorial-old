# RMSD analysis
- Group for least square fit: Backbone
- Group for RMSD calculation: Backbone

```python
!echo "4 4" | gmx rms -s traj-fit-nodt.pdb -f traj-fit-nodt-500ns.xtc -o rmsf.xvg
```

```python
# Load the data from the first file
data = np.loadtxt('trajrmsd.dat',skiprows=2)
time = data[:, 0] 
rmsd = data[:, 1]*0.05

# Create the plot
plt.figure(figsize=(10, 6))
plt.title("RMSD")
plt.plot(time, rmsd)
plt.ylim(bottom=0)
plt.xlabel("RMSD (nm)")
plt.ylabel("Time (ns)")
plt.savefig('rmsd_plot.png')
plt.grid(True)
plt.tight_layout()
plt.show()
```
